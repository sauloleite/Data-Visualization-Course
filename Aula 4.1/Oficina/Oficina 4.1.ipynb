```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
plt.figure(figsize=(20,16))
plt.title('Distribuição do valor da passagem de acordo com a cidade')
sns.violinplot(x="embark_town", y="fare", data=titanic)
plt.show()
```

```python
sns.catplot(x="embark_town", y="fare", hue = 'sex', col = 'survived', kind = 'violin', data=titanic)
'''O preço das passagens dos que sobreviveram é maior do que os não sobreviventes. Desse modo, mostrando que os de classe social mais privilegiada foram os primeiros a abandonarem o navio. Sendo estes homens e mulheres da grande maioria de Cherbourg.'''
```
