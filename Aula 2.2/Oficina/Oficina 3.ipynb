```python
'''1. Importe as bibliotecas pandas e matplotlib, usando a palavra mágica inline para exibir as visualizações no notebook; e carregue os dados, que estão disponibilizados neste link:dados, em um DataFrame. Assim, você pode carregar os dados pela URI ou salvar em sua máquina e carregá-los localmente;'''
#É preciso importar e configurar as bibliotecas que serão utilizadas.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
url_videogames = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt'
videogames = pd.read_csv(url_videogames)
videogames.head()
```

```python
'''2. Crie uma visualização que mostre a proporção de vendas global entre gêneros de jogos da empresa Nintendo'''
videogames = videogames[videogames['Publisher']=='Nintendo'] #Pegando dados somente da empresa Nintendo
generos = videogames.groupby(by='Genre').sum()
plt.figure(figsize=(15,8))
plt.pie(x=generos['Global_Sales'], labels = generos.index, autopct='%1.1f%%',startangle=90, pctdistance=0.85)
plt.title('Proporção de vendas global entre gêneros de jogos da empresa Nintendo')
plt.show()
```

```python
#3. Crie uma visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015.
qtds = [] #Criando uma lista vazia para armazenar as quantidades
videogames = pd.read_csv(url_videogames) #Retomando dados do dataset original
videogames = videogames[videogames['Year']>=2015] #Definindo somente dados a partir do ano de 2015
plataformas = sorted(set(videogames['Platform'])) #Criando uma lista com as plataformas
for i in plataformas: #Loop para verificar a quantidade de jogos de cada plataforma
    qtds.append(len((videogames[videogames['Platform']==i]))) #Adicionando a quantidade a lista "qtds"
#Plotando...
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15,5))
plt.bar(plataformas,qtds)
plt.title('Proporção de quantidade de jogos por plataforma a partir do ano de 2015')
plt.show()
```
