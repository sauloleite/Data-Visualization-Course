```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relao entre a acelerao e o consumo de combustvel')
sns.scatterplot(x="acceleration", y="mpg", hue = 'origin', data=mpg)
plt.show()
'''Sabendo disso, existe uma correlao entre a acelerao e o consumo? De que tipo? A correlao  a mesma para os diferentes pases de origem? R:  possvel observar que a o consumo costuma ser maior quando a acelerao atinge os valores entre 14 e 20. Desse modo, h alta correlao Positiva entre os dois. Como mostra as cores no grfico, a correlao no  a mesma entre os pases de origens'''
```

```python
plt.figure(figsize=(8,5))
plt.title('Relao entre a acelerao e o consumo de combustvel')
sns.scatterplot(x="acceleration", y="mpg", hue = 'model_year', data=mpg)
plt.show()
'''Existe alguma relao entre as trs variveis? R: Sim,  possvel observar que o consumo por acelerao foi aumentando conforme os modelos dos carros foram ficando mais novos.'''
```
