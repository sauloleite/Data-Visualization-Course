```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre a aceleração e o consumo de combustível')
sns.scatterplot(x="acceleration", y="mpg", hue = 'origin', data=mpg)
plt.show()
'''Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem? R: É possível observar que a o consumo costuma ser maior quando a aceleração atinge os valores entre 14 e 20. Desse modo, há alta correlação Positiva entre os dois. Como mostra as cores no gráfico, a correlação não é a mesma entre os países de origens'''
```

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre a aceleração e o consumo de combustível')
sns.scatterplot(x="acceleration", y="mpg", hue = 'model_year', data=mpg)
plt.show()
'''Existe alguma relação entre as três variáveis? R: Sim, é possível observar que o consumo por aceleração foi aumentando conforme os modelos dos carros foram ficando mais novos.'''
```
