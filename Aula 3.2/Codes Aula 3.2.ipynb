```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relao entre valor da conta e valor da gorjeta')
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relao entre valor da conta e valor da gorjeta')
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relao entre valor da conta e valor da gorjeta pelo gnero')
sns.scatterplot(x="total_bill", y="tip", hue = 'sex', data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relao entre valor da conta e valor da gorjeta pelo dia da semana')
sns.scatterplot(x="total_bill", y="tip", hue = 'day', data=tips)
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", col = 'day', data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue = 'sex', data=tips)
```

```python
plt.figure(figsize=(8,5))
plt.title('Relao entre valor da conta, valor da gorjeta e tamanho da mesa')
sns.scatterplot(x="total_bill", y="tip", hue = 'size', data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relao entre valor da conta, valor da gorjeta e tamanho da mesa')
sns.scatterplot(x="total_bill", y="tip", size = 'size', sizes=(15, 200), data=tips)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Gorjetas')
sns.scatterplot(x="total_bill", y="tip", size = 'size', hue = 'sex', style = 'smoker', sizes=(15, 200), data=tips)
plt.show()
```
