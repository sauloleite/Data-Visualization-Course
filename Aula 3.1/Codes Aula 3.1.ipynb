```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
houses_to_rent_v3 = 'https://leadfortaleza.com.br/ead2pcd/conteudo/tmp/myopenolat_1_102915070706483/aula/arquivos/dados.txt'
df = pd.read_csv(houses_to_rent_v3)
df.head()
```

```python
df['area'].describe()
```

```python
plt.figure(figsize=(10, 5))
plt.style.use('seaborn-darkgrid')
plt.title('Distribui巽達o de 叩reas dos im坦veis')
plt.hist(df['area'])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Distribui巽達o de 叩reas dos im坦veis')
plt.hist(df['area'],bins=25)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Distribui巽達o de 叩reas dos im坦veis')
sns.distplot(df['area'])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Distribui巽達o de 叩reas dos im坦veis')
sns.kdeplot(df['area'], shade=True)
plt.show()
```
