```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
dadosimobiliarios = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro05/oficina/arquivos/dadosoficina05.txt'
df = pd.read_csv(dadosimobiliarios)
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Distribuição de áreas dos imóveis')
sns.distplot(df['rent amount (R$)'])
plt.show()
'''Qual faixa de valores de aluguel é mais comum? R: A faixa de valores de aluguel mais comum é cerca de R$ 2700.'''
```

```python
plt.figure(figsize=(10, 5))
plt.title('Distribuição de áreas dos imóveis')
sns.distplot(df['fire insurance (R$)'])
plt.show()
'''Existem valores incomuns? R: A partir de 50 reais, os valores começam a ficar incomuns.'''
```
